```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Import OS and JSON Modules
import os, json
```

```python
filename = ('Data/superhero_info.csv')
df = pd.read_csv(filename)
df.head()
```

```python
filename2 = ('Data/superhero_powers.csv')
df2 = pd.read_csv(filename2)
df2.head()
```

## Info Dataframe

```python
df.info()
```

```python
df['Hero|Publisher'].value_counts()
```

**Split Hero|Publisher**

```python
df['Hero|Publisher'].str.split('|', expand=True)
```

```python
# save two new columns into dataframe
df[['Hero','Publisher']] = df['Hero|Publisher'].str.split('|', expand=True)
df.head()
```

```python
# drop original column
df = df.drop(columns = ['Hero|Publisher'])
df.head()
```

**Make Dictionary**

```python
coord = df.loc[0,"Measurements"]
print(type(coord))
coord
```

```python
# Replace single quotes
coord = coord.replace("'",'"')
coord
```

```python
# Use json.loads
fixed_coord = json.loads(coord)
print(type(fixed_coord))
fixed_coord
```

```python
# use .str.replace to replace all single quotes
df['Measurements'] = df['Measurements'].str.replace("'",'"')

## Apple json.loads to full column
df['Measurements'] = df['Measurements'].apply(json.loads)
df['Measurements'].head()
```

```python
## check a single value after transformation
test_coord = df.loc[0, 'Measurements']
print(type(test_coord))
test_coord
```

**Splitting Dictionary**

```python
height_weight = df['Measurements'].apply(pd.Series)
height_weight
```

```python
# Add to dataframe and drop original column

df = pd.concat((df, height_weight), axis = 1)

df = df.drop(columns=['Measurements'])
df.info()
df.head()
```

**Fixing Height Column**

```python
## adding expand=True
df['Height'].str.split(' ',expand=True)
```

```python
df[['Height_cm', 'Units']] = df['Height'].str.split(' ',expand=True)
df.head()
```

```python
df = df.drop(columns = ['Units', 'Height'])
df.head()
```

**Fixing Weight Column**

```python
## adding expand=True
df['Weight'].str.split(' ',expand=True)
```

```python
df[['Weight_kg', 'Units2']] = df['Weight'].str.split(' ',expand=True)
df.head()
```

```python
df = df.drop(columns = ['Units2', 'Weight'])
df.head(2)
```

```python
df.info()
```

**Converting Height and Weight to Float**

```python
df['Height_cm'] = df['Height_cm'].astype(float)
print(df.dtypes)
```

```python
df['Weight_kg'] = df['Weight_kg'].astype(float)
print(df.dtypes)
```

```python
df.to_csv('advanced2_tf_data_pt1.csv', index = False)
```

## Powers Dataframe

```python
df2.info()
df2.head()
```

```python
df2['Powers'].value_counts()
```

```python
## adding expand=True
df2['powers_split'] = df2['Powers'].str.split(',',expand=False)
```

```python
df2.head(2)
```

```python
## exploding the column of lists

exploded = df2.explode('powers_split')
exploded[['hero_names','Powers', 'powers_split']].head(5)
```

```python
## saving the unique values from the exploded column
cols_to_make = exploded['powers_split'].dropna().unique()
cols_to_make
```

```python
for col in cols_to_make:
    df2[col] = df2['Powers'].str.contains(col)
df2.head()
```

```python
# Drop Powers Column

df2 = df2.drop(columns=['Powers', 'powers_split'])
df2.to_csv('advanced_tf_data_pt1.csv', index = False)
```
